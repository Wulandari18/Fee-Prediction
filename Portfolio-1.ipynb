Data in 'Kelas Pekerja' and 'Pekerjaan' have not number data (? sign), so we must to replace or drop this data.

```python
dataset = dataset.replace('?', np.nan)
```

```python
dataset.isnull().sum()
```

There are two columns with missing value, 'Kelas Pekerja' and 'Pekerjaan'. Lets change the missing value with the modus in each column. 

```python
dataset['Kelas Pekerja'] = dataset['Kelas Pekerja'].fillna('Wiraswasta')
dataset['Pekerjaan'] = dataset['Pekerjaan'].fillna('Spesialis')
```

We don't have any information about category 1st - 4th, 5th - 6th, 12th, 9th, 7th - 8th, 10th, and 11th in 'Pendidikan' data. So we may delete this data to get more insight.

```python
index_names = dataset[(dataset['Pendidikan'] == '1st-4th') | (dataset['Pendidikan'] == '5th-6th') | 
                      (dataset['Pendidikan'] == '12th') | (dataset['Pendidikan'] == '9th') | 
                        (dataset['Pendidikan'] == '7th-8th') | (dataset['Pendidikan'] == '10th') | 
                      (dataset['Pendidikan'] == '11th')].index

dataset.drop(index_names, inplace = True)
display(dataset['Pendidikan'].value_counts()) 
```

# Encoding of non-numerical values

```python
# Working Class, Married Status, and Job
dataset = pd.get_dummies(dataset, columns = ['Kelas Pekerja', 'Status Perkawinan', 'Pekerjaan'])

# Education
dict_pen = {
    'SD' : 0,
    'SMP' : 1,
    'SMA' : 2,
    'D3' : 3,
    'D4' : 4,
    'Sekolah Professional' : 5,
    'Pendidikan Tinggi' : 6,
    'Sarjana' : 7,
    'Master' : 8,
    'Doktor' : 9,
}

dataset['Pendidikan'] = dataset['Pendidikan'].replace(dict_pen)

# Gender
dict_gender = {
    'Perempuan' : 0,
    'Laki2' : 1,
}

dataset['Jenis Kelamin'] = dataset['Jenis Kelamin'].replace(dict_gender)
```

```python
# Fee
dict_gaji = {
    '<=7jt' : 0,
    '>7jt' : 1,
}

dataset['Gaji'] = dataset['Gaji'].replace(dict_gaji)

dataset.head()
```

### Model Selection

#### Import Machine Learning Model

```python
from sklearn.linear_model import LogisticRegression

from sklearn.tree import DecisionTreeClassifier

from sklearn.neighbors import KNeighborsClassifier

from sklearn.svm import SVC
```

#### Import Machine Learning Library

```python
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score

from sklearn.metrics import classification_report

from sklearn.model_selection import KFold

from sklearn.model_selection import cross_validate, cross_val_score

from sklearn import metrics
```

```python
def model_selection(model, X_train, X_test, y_train, y_test):
    
    model.fit(X_train, y_train)
    score = model.score(X_train, y_train)
    predictions = model.predict(X_test)
    accuracy = accuracy_score(predictions, y_test)
    
    return (score, accuracy, predictions)   
```

### Tunning Parameters

```python
from  sklearn.model_selection import RandomizedSearchCV
```

```python
model = DecisionTreeClassifier()

param_grid = {'max_depth' : np.arange(4, 20), 'min_samples_leaf' : np.arange(3, 20)}
dtc_model = RandomizedSearchCV(model, param_grid, scoring = 'roc_auc', cv = 5)
dtc_model.fit(X_train, y_train)

print("\n Best Score is ")
print(dtc_model.best_score_)

print("\n Best Estimator is ")
print(dtc_model.best_estimator_)

print("\n Best Parametes are")
print(dtc_model.best_params_)
```

```python
model = KNeighborsClassifier()

param_grid = {'n_neighbors' : np.arange(2, 50), 'weights' : ['distance', 'uniform']}
knn_model = RandomizedSearchCV(model, param_grid, scoring = 'roc_auc', cv = 10)
knn_model.fit(X_train, y_train)

print("\n Best Score is ")
print(knn_model.best_score_)

print("\n Best Estimator is ")
print(knn_model.best_estimator_)

print("\n Best Parametes are")
print(knn_model.best_params_)
```

```python
models_list = {
    "LogisticRegression" :  LogisticRegression(),
    "DecisionTreeClassifier" :  DecisionTreeClassifier(max_depth = 10, min_samples_leaf = 17),
    "KNeighborsClassifier" : KNeighborsClassifier(n_neighbors = 39, weights = 'uniform'),
    "SVC" :  SVC(),
}
```

```python
# Let's Define the function for confision metric Graphs

def cm_metrix_graph(cm):
    
    sns.heatmap(cm,annot=True,fmt="d")
    plt.show()
```

```python
df_prediction = []
confusion_matrixs = []
df_prediction_cols = [ 'model_name', 'score', 'accuracy_score' , "accuracy_percentage"]

for name, model in zip(list(models_list.keys()), list(models_list.values())):
    
    (score, accuracy, predictions) = model_selection(model, X_train, X_test, y_train, y_test )
    
    print("\n\nClassification Report of '"+ str(name), "'\n")
    
    print(classification_report(y_test, predictions))

    df_prediction.append([name, score, accuracy, "{0:.2%}".format(accuracy)])
    
    # For Showing Metrics
    confusion_matrixs.append(confusion_matrix(y_test, predictions))
    
        
df_pred = pd.DataFrame(df_prediction, columns=df_prediction_cols)
```

```python
plt.figure(figsize=(16, 4))
# plt.title("Confusion Metric Graph")


for index, cm in enumerate(confusion_matrixs):

    plt.subplot(1, len(confusion_matrixs), index+1)
    
    #plt.xlabel("Negative Positive")
    #plt.ylabel("True Positive")

    
    
    # Show The Metrics Graph    
    cm_metrix_graph(cm) # Call the Confusion Metrics Graph
    plt.tight_layout(pad=True)
```
